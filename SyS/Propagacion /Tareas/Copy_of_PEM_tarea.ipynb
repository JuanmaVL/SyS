```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros
k =  np.pi / 1      # Número de onda
omega = 2 * np.pi * 1  # Frecuencia angular
z = np.linspace(0, 10, 50)
x = np.zeros_like(z)
y = np.zeros_like(z)

# Tiempo (puedes animar cambiándolo)
t = 0

# Campos
E0 = 1
B0 = 1

Ex = E0 * np.cos(k * z - omega * t)
By = B0 * np.cos(k * z - omega * t)

# Creamos la figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar los vectores E (en x) y B (en y), propagando en z
ax.quiver(x, y, z, Ex, 0*y, 0*z, color='r',arrow_length_ratio=0.01, label='Campo E (x)')
ax.quiver(x, y, z, 0*x, By, 0*z, color='b', arrow_length_ratio=0.01,label='Campo B (y)')

# Estética
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([0, 10])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z (propagación)')
ax.set_title('Onda Electromagnética (E rojo, B azul)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
f = 2.4e9                        # frecuencia [Hz]
c = 3e8                          # velocidad luz en el vacío
er = 4.5                         # permitividad relativa (FR4)
tan_delta = 0.02                # factor de pérdidas
lambda_d = c / (f * np.sqrt(er))  # longitud de onda en el medio
beta = 2 * np.pi / lambda_d
alpha = np.pi * er * tan_delta / lambda_d

# Eje espacial (z)
z = np.linspace(0, 0.5, 1000)  # hasta 5 cm
E0 = 1                          # amplitud
t = 0                           # tiempo fijo para ver forma espacial

# Campo eléctrico
E = E0 * np.exp(-alpha * z) * np.cos(beta * z)

# Gráfica
plt.figure(figsize=(10,4))
plt.plot(z*100, E, label='Campo E(z) con pérdidas (FR4)', color='orange')
plt.title('Forma de onda del Campo Eléctrico a 2.4 GHz en FR4 (modelo con pérdidas)')
plt.xlabel('Distancia z [cm]')
plt.ylabel('Amplitud del Campo E [u.a.]')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define los parámetros de la onda
z = np.linspace(0, 10, 100)  # Rango de valores para z
t = 0  # Tiempo (puedes cambiarlo para ver la evolución de la onda)
w = 1  # Frecuencia angular

# Define una lista de valores para k
k_values = [1, 2, 3, 4]

# Crea una figura para la visualización
plt.figure(figsize=(8, 6))

# --- Primera ecuación (cos(kz-wt)ax + cos(kz-wt)ay) ---
for k in k_values:
    Ex = np.cos(k * z - w * t)
    Ey = np.cos(k * z - w * t)
    E_total = Ex + Ey  # Suma algebraica
    plt.plot(z, E_total, label=f'k = {k} (cos + cos)')

# --- Segunda ecuación (cos(kz-wt)ax + sen(kz-wt)ay) ---
for k in k_values:
    Ex = np.cos(k * z - w * t)
    Ey = np.sin(k * z - w * t)
    E_total = Ex + Ey  # Suma algebraica
    plt.plot(z, E_total, label=f'k = {k} (cos + sen)')

# Configura la gráfica
plt.xlabel('z')
plt.ylabel('E(z, t) Suma Algebraica')
plt.title('Onda Cosenoidal y Senoidal con Diferentes Valores de k')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda
z = np.linspace(0, 10, 20)  # Rango de valores para z
t = 0  # Tiempo
w = 1  # Frecuencia angular
k = 1  # Número de onda

# Calcula las componentes para la primera ecuación (cos + cos)
Ex1 = np.cos(k * z - w * t)
Ey1 = np.cos(k * z - w * t)

# Calcula las componentes para la segunda ecuación (cos + sen)
Ex2 = np.cos(k * z - w * t)
Ey2 = np.sin(k * z - w * t)

# Crea la figura y los ejes
fig, ax = plt.subplots()

# Representación vectorial para la primera ecuación (color azul)
ax.quiver(z, np.zeros_like(z), Ex1, Ey1, angles='xy', scale_units='xy', scale=1, color='blue', label='cos(kz-wt)ax + cos(kz-wt)ay')

# Representación vectorial para la segunda ecuación (color rojo)
ax.quiver(z, np.zeros_like(z), Ex2, Ey2, angles='xy', scale_units='xy', scale=1, color='red', label='cos(kz-wt)ax + sen(kz-wt)ay')

# Configura la gráfica
ax.set_xlabel('z')
ax.set_ylabel('Amplitud')
ax.set_title('Representación vectorial de ambas ecuaciones')
ax.grid(True)
ax.legend()  # Agrega una leyenda para distinguir las ondas
plt.show()
```
