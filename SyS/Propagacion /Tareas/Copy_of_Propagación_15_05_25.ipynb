```python
#EJERCICIO 2.40


import numpy as np
import matplotlib.pyplot as plt

# Índices de refracción
n1 = 1.48  # Vidrio (núcleo de la fibra)
n2 = 1.00  # Aire (defecto)

# Cálculo del ángulo crítico en radianes y grados
theta_crit_rad = np.arcsin(n2 / n1)
theta_crit_deg = np.degrees(theta_crit_rad)

# Ángulo de incidencia dado
theta_inc_deg = 80
total_internal_reflection = theta_inc_deg > theta_crit_deg

# Resultados
print(f"Índice de refracción del vidrio (n1): {n1}")
print(f"Índice de refracción del aire (n2): {n2}")
print(f"Ángulo crítico: {theta_crit_deg:.2f}°")
print(f"Ángulo de incidencia: {theta_inc_deg}°")
print("¿Ocurre reflexión interna total? ", "Sí" if total_internal_reflection else "No")

# Graficar
plt.figure(figsize=(8, 5))
plt.axhline(theta_crit_deg, color='red', linestyle='--', label=f'Ángulo Crítico ({theta_crit_deg:.2f}°)')
plt.axhline(theta_inc_deg, color='blue', linestyle='-', label=f'Ángulo de Incidencia ({theta_inc_deg}°)')

plt.title('Comparación entre Ángulo de Incidencia y Ángulo Crítico')
plt.ylabel('Ángulo (grados)')
plt.ylim(0, 90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#EJERCICIO 2.39

import numpy as np
import matplotlib.pyplot as plt

# Índices de refracción
n1 = 1.48  # Núcleo de la fibra
n2 = 1.45  # Revestimiento (cladding)

# Cálculo del ángulo crítico en radianes y grados
theta_crit_rad = np.arcsin(n2 / n1)
theta_crit_deg = np.degrees(theta_crit_rad)

# Imprimir resultado
print(f"Índice de refracción del núcleo (n1): {n1}")
print(f"Índice de refracción del revestimiento (n2): {n2}")
print(f"Ángulo crítico en la interfaz núcleo-revestimiento: {theta_crit_deg:.2f}°")

# Gráfica
plt.figure(figsize=(8, 5))
plt.axhline(theta_crit_deg, color='green', linestyle='--', label=f'Ángulo Crítico ({theta_crit_deg:.2f}°)')
plt.title('Ángulo Crítico para Reflexión Interna Total en Fibra Óptica')
plt.ylabel('Ángulo (grados)')
plt.ylim(0, 90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#EJERCICIO 2.41

import numpy as np
import matplotlib.pyplot as plt

# Índices de refracción
n1 = 1.00  # Aire
n2 = 1.58  # Policarbonato

# Ángulo de Brewster en radianes y grados
theta_b_rad = np.arctan(n2 / n1)
theta_b_deg = np.degrees(theta_b_rad)

# Imprimir resultado
print(f"Índice de refracción del aire (n1): {n1}")
print(f"Índice de refracción del policarbonato (n2): {n2}")
print(f"Ángulo de Brewster: {theta_b_deg:.2f}°")

# Gráfica
plt.figure(figsize=(8, 5))
plt.axhline(theta_b_deg, color='orange', linestyle='--', label=f'Ángulo de Brewster ({theta_b_deg:.2f}°)')
plt.title('Ángulo de Brewster en interfaz aire-policarbonato')
plt.ylabel('Ángulo (grados)')
plt.ylim(0, 90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#EJERCICIO 2.42

import numpy as np

# Parámetros
Z1 = 1.00  # Impedancia del aire
Z2 = 0.40  # Impedancia del foam
theta_i_deg = 30
theta_i_rad = np.radians(theta_i_deg)

# Usamos aproximación de n1 = Z1, n2 = Z2 para Snell
sin_theta_t = (Z1 / Z2) * np.sin(theta_i_rad)

# Manejo de total reflection
if abs(sin_theta_t) > 1:
    print("Reflexión total: no hay transmisión.")
else:
    theta_t_rad = np.arcsin(sin_theta_t)
    R = ((Z2 * np.cos(theta_i_rad) - Z1 * np.cos(theta_t_rad)) /
         (Z2 * np.cos(theta_i_rad) + Z1 * np.cos(theta_t_rad)))**2
    T = 1 - R

    print(f"Ángulo de incidencia: {theta_i_deg}°")
    print(f"Ángulo de transmisión: {np.degrees(theta_t_rad):.2f}°")
    print(f"Coeficiente de reflexión (intensidad): {R:.4f}")
    print(f"Coeficiente de transmisión (intensidad): {T:.4f}")
```
