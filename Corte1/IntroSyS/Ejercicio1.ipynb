```python
import numpy as np
import matplotlib.pyplot as plt
# definir parametros de la seal
A1 = 20
A2 = 50
w1 = 100
w2 = 200
T1 = 2*np.pi/w1
T2 = 2*np.pi/w2
F1 = w1/(2*np.pi)
F2 = w2/(2*np.pi)
# definir frecuencia de muestreo nyquist
Fs = 50*max(F1,F2)
# periodo seal cuaciperiodica T = mcm (T1,T2)
T = np.pi/50
# definir vector de tiempo
ti = -2*T
tf = 3*T
tv = np.arange(ti,tf,1/Fs)

# construir la seal
phi2 = 5
x = A1*np.cos(w1*tv) - A2*np.sin(w2*tv+phi2)
xc = A1*np.cos(w1*tv)
xs = -A2*np.sin(w2*tv+phi2)
plt.plot(tv,x,'r', label='$x(t)')
plt.plot(tv,xc,'b',label='$200\cos(100t)$')
plt.plot(tv,xs,'g',label='$-50\cos(200t+5)$')
plt.legend()
plt.xlabel('t')
plt.ylabel('$x(t)')
plt.grid(True)
plt.show()

print(f'FI={F1} Hz F2={Fs} Hz')
```

```python
import numpy as np
import matplotlib.pyplot as plt
# definir parametros de la seal
A1 = 1
A2 = 1
w1 = 1/3
w2 = 1/4
T1 = 2*np.pi/w1
T2 = 2*np.pi/w2
F1 = w1/(2*np.pi)
F2 = w2/(2*np.pi)
# definir frecuencia de muestreo nyquist
Fs = 10*max(F1,F2)
# periodo seal cuaciperiodica T = mcm (T1,T2)
T = 24*np.pi
# definir vector de tiempo
ti = -1*T
tf = 1*T
tv = np.arange(ti,tf,1/Fs)

# construir la seal
x = A1*np.cos(w1*tv) + A2*np.cos(w2*tv)
xc = A1*np.cos(w1*tv)
xs = -A2*np.cos(w2*tv+phi2)

plt.stem(tv,x)
plt.plot(tv,x,'r', label='$x(t)')
plt.plot(tv,xc,'b',label='$\cos(t/3)$')
plt.plot(tv,xs,'g',label='$\cos(t/4)$')
plt.legend()
plt.xlabel('t')
plt.ylabel('$x(t)')
plt.grid(True)
plt.show()

print(f'F1={F1} Hz F2={Fs} Hz')
```

```python
y_min = 0 # valor min despus de cero y pendiente
y_max = 10 # valor max despus de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
```

```python
ymin = -3.3
ymax = 3.3
xcp = my_ceropen(x,ymin,ymax)
plt.plot(tv,xcp,c='r',label='cero_pendiente')
plt.legend()
plt.grid(True)
plt.show()

```

```python
from scipy.spatial.distance import cdist
nbits = 4
num_est = 2**nbits
vq = np.linspace(ymin,ymax,num_est) # vector de cauntizacion
```

```python
#funcin cuantizacin punto a punto

def my_cuantizador(yn, vq) : #yn punto a #cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamao vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1))#distancia yn a vector estados, reshape(-1,1) asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posicin distancia min
  return vq[ind]
```

```python
#recorrer vector cero y spam discretizado para cuantizar
for i in range(xcp.shape[0]):
   if i==0: #primera muestra
     yq = my_cuantizador(xcp[i],vq)
   else:
     yq = np.append(yq,my_cuantizador(xcp[i],vq)) #apilar datos cuantizando si i>0
```

```python
#graficar seales

plt.plot(tv,xcp,c='r', label='xcp')#cs continua

plt.stem(tv,xcp, label='xcp_dis')#cs discreta

#estados de cuantizacin

for v in vq:
  plt.plot(tv,v*np.ones(len(tv)))

# graficar seal Digital (discreta + cuantizada)

plt.plot(tv,yq,c='y',marker='o',label='ydig', linewidth=2)

plt.legend()
plt.grid(True)
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.show()
```
