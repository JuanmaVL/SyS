```python
import numpy as np
import matplotlib.pyplot as plt

# Pedir al usuario la amplitud y el rango de x
A = float(input("Ingresa la amplitud (A): "))
x_inicio = float(input("Ingresa el valor inicial de x: "))
x_fin = float(input("Ingresa el valor final de x: "))

# Crear el vector x
x = np.linspace(x_inicio, x_fin, 100)

# Crear el vector y (fase) - FIJO de 0 a 2pi
y = np.linspace(0, 2 * np.pi, 100)

# Calcular z
z = A * np.sin(x + y)

# Crear la gráfica
plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("z")
plt.title("Gráfica de z = Asen(x + y)")
plt.grid(True)
plt.show()
```
