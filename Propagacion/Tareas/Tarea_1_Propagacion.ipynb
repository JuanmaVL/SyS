```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la frecuencia
f = 150e6  # 150 MHz

# Definir la velocidad de la luz en el vacío
c = 3e8  # m/s

# Definir las permitividades relativas de los medios
er_papel = 3.0  # Papel
er_fr4 = 4.4  # FR4
er_cobre = 1.0  # Cobre (aproximado)

# Definir las permiabilidades relativas de los medios
ur_papel = 1.0  # Papel
ur_fr4 = 1.0  # FR4
ur_cobre = 1.0  # Cobre (aproximado)

# Calcular las velocidades de fase en cada medio
vp_papel = c / np.sqrt(er_papel*ur_papel)
vp_fr4 = c / np.sqrt(er_fr4*ur_fr4)
vp_cobre = c / np.sqrt(er_cobre*ur_cobre)

# Definir el rango de x
x = np.linspace(0, 3, 3000)  # 3 metro

# Calcular la longitud de onda en cada medio
lambda_papel = vp_papel / f
lambda_fr4 = vp_fr4 / f
lambda_cobre = vp_cobre / f

# Calcular la fase en cada medio para t=0
fase_papel = 2 * np.pi * x / lambda_papel
fase_fr4 = 2 * np.pi * x / lambda_fr4
fase_cobre = 2 * np.pi * x / lambda_cobre

# Calcular la amplitud de la onda en cada medio (asumiendo la misma amplitud)
A = 1

# Calcular la onda en cada medio
y_papel = A * np.sin(fase_papel)
y_fr4 = A * np.sin(fase_fr4)
y_cobre = A * np.sin(fase_cobre)

# Graficar las ondas superpuestas
plt.plot(x, y_papel, label='Papel')
plt.plot(x, y_fr4, label='FR4')
plt.plot(x, y_cobre, label='Cobre')

# Configurar la gráfica
plt.xlabel('Distancia (m)')
plt.ylabel('Amplitud')
plt.title('Onda viajera a 150 MHz en diferentes medios (t=0)')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid coordinates
x, y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))

# Calculate vector components (example: a simple radial field)
u = x
v = y

plt.quiver(x, y, u, v)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field')
plt.show()
plt.quiver(x, y, u, v, color='red', scale=10, width=0.005, headwidth=3)
```
