```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
import ipywidgets as widgets
from ipywidgets import interactive

# Función para actualizar el Lugar de las Raíces
def analizar_LGR(K):
    if K == 0:
        # Mostrar mensaje para K = 0
        print("Para K = 0, el sistema es trivial (G(s) = 0) y no tiene Lugar de las Raíces.")
        return

    # Definir la función de transferencia G(s) = K*(s + 3) / (s^2 + 6s + 11)
    s = ctrl.TransferFunction([1, 3], [1, 6, 11])  # (s + 3) / (s^2 + 6s + 11)
    G = K * s  # Multiplicar por K

    # Crear la figura para el Lugar de las Raíces
    fig, ax = plt.subplots(figsize=(8, 6))

# Marcar los ejes principales
    ax.axhline(0, color='black',linewidth=1)  # Eje real
    ax.axvline(0, color='black',linewidth=1)  # Eje imaginario

    # Lugar de las Raíces
    ctrl.root_locus(G, ax=ax)
    ax.set_title(f'LGR para K = {K}')

    plt.tight_layout()  # Ajustar los márgenes automáticamente
    plt.show()

# Crear un control deslizante interactivo para modificar K
interactive_plot = interactive(analizar_LGR, K=widgets.FloatSlider(value=1.0, min=-10.0, max=10.0, step=0.1))
output = interactive_plot.children[-1]
output.layout.height = '400px'
interactive_plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
from ipywidgets import interactive, FloatSlider, VBox

# Función para analizar el sistema con un polo añadido en el origen
def analizar_LGR(K_max):
    # Definir la función de transferencia original: G(s) = K * (s + 3) / (s^2 + 6s + 11)
    num = [1, 3]  # Numerador (s + 3)
    den = [1, 6, 11]  # Denominador (s^2 + 6s + 11)

    # Añadir un polo en el origen (multiplicar el denominador por s)
    den_modificado = np.convolve(den, [1, 0])  # (s^3 + 6s^2 + 11s)

    # Crear la función de transferencia con el polo en el origen
    G = ctrl.TransferFunction(num, den_modificado)

    # Graficar el Lugar de las Raíces
    fig, ax = plt.subplots(figsize=(8, 6))
    ctrl.root_locus(G, kvect=np.linspace(0, K_max, 500), ax=ax)  # Aumentar el rango de K

    # Marcar los ejes principales
    ax.axhline(0, color='black',linewidth=1)  # Eje real
    ax.axvline(0, color='black',linewidth=1)  # Eje imaginario

    ax.set_title(f'LGR del sistema con polo en el origen (K hasta {K_max})')
    ax.grid(True)
    plt.tight_layout()
    plt.show()


# Crear un control deslizante interactivo para modificar K_max
interactive_plot = interactive(
    analizar_LGR,
    K_max=FloatSlider(value=10.0, min=1.0, max=50.0, step=1.0, description="Ganancia K_max")
)

# Ajustar el tamaño del gráfico interactivo
output = interactive_plot.children[-1]
output.layout.height = '400px'

# Mostrar el control deslizante interactivo
VBox([interactive_plot])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
from ipywidgets import interactive, FloatSlider, VBox

def analizar_LGR_con_polos(K_max):
    """
    Graficar el Lugar de las Raíces para el sistema con dos polos complejos conjugados adicionales.

    Parámetros:
    - K_max: Valor máximo de la ganancia K.
    """
    # Numerador y denominador de la función de transferencia original
    num = [1, 3]  # Numerador: (s + 3)
    den = [1, 6, 11]  # Denominador: (s^2 + 6s + 11)

    # Añadir dos polos complejos conjugados (frecuencia natural 2 rad/s y un damping de 0.5)
    polos_complejos = np.poly([-1 + 1j * np.sqrt(3), -1 - 1j * np.sqrt(3)])  # (s + 1)^2 + 3
    den_modificado = np.convolve(den, polos_complejos)  # Multiplicar denominadores

    # Crear la función de transferencia modificada
    G = ctrl.TransferFunction(num, den_modificado)



    # Generar y graficar el Lugar de las Raíces
    fig, ax = plt.subplots(figsize=(8, 6))

    ctrl.root_locus(G, kvect=np.linspace(0, K_max, 1500), ax=ax)  # Variar K desde 0 hasta K_max

    ax.set_title(f'LGR del sistema con polos complejos conjugados (K hasta {K_max})')
    ax.set_xlabel('Parte Real')
    ax.set_ylabel('Parte Imaginaria')
    ax.grid(True)
    # Marcar los ejes principales
    ax.axhline(0, color='black',linewidth=1)  # Eje real
    ax.axvline(0, color='black',linewidth=1)  # Eje imaginario
    plt.tight_layout()
    plt.show()

# Crear un control deslizante interactivo para K_max
interactive_plot = interactive(
    analizar_LGR_con_polos,
    K_max=FloatSlider(value=10.0, min=1.0, max=100.0, step=1.0, description="Ganancia K_max")
)

# Ajustar el tamaño del gráfico interactivo
output = interactive_plot.children[-1]
output.layout.height = '400px'

# Mostrar el control deslizante interactivo
VBox([interactive_plot])
```
