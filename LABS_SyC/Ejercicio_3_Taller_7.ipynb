#***CASO A***

```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
from scipy.integrate import odeint

A = np.array([[0, 1], [-8, -6]])
B = np.array([[0], [1]])
x0 = np.array([[1], [0]])
t = np.linspace(0, 10, 1000)

def system_step(x, t, A, B):
    u = 1  # Step input
    dxdt = A @ x + (B * u).flatten()
    return dxdt

# Solve the system with initial condition x0 and step input
x = odeint(system_step, x0.flatten(), t, args=(A, B))

plt.plot(t, x[:, 0], label="x1(t)")
plt.plot(t, x[:, 1], label="x2(t)")
plt.title("Respuesta al escal√≥n")
plt.legend()
plt.grid()
plt.show()
```

#**CASO B**

```python
from scipy.integrate import odeint

def system(x, t):
    u = np.exp(-2*t)
    dxdt = A @ x + (B * u).flatten()
    return dxdt

x = odeint(system, x0.flatten(), t)

plt.plot(t, x[:, 0], label="x1(t)")
plt.plot(t, x[:, 1], label="x2(t)")
plt.title("Respuesta con entrada u(t)=e^{-2t}")
plt.legend()
plt.grid()
plt.show()
```
